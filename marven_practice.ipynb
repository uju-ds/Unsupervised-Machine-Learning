```python
new_data= (ent_data.pivot(index='name', 
                          columns = 'entertainment',
                          values='hours_per_week')
           .fillna(0).reset_index())
new_data
```

## Column Preparation

```python
import numpy as np

new_data['video_games_lovers'] = np.where(new_data['video_games'] > 7,1,0)
new_data
```

## Feature Engineering

```python
new_data['total_entertainment'] = new_data['books'] + new_data['movies']+ new_data['tv_shows'] + new_data['video_games']
new_data
```

```python
new_data['pct_screen'] = (new_data['movies']+ new_data['tv_shows'] + new_data['video_games'])/new_data['total_entertainment']
new_data
```

```python
name= new_data['name']
name
```

```python
data_drop=new_data.drop(columns=['name'])
data_drop
```

## Feature selection

```python
data_series= data_drop[['video_games_lovers','total_entertainment','pct_screen']]
data_series
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
standardized = std_scaler.fit_transform(data_series)

model_data = pd.DataFrame(standardized, columns=data_series.columns)
model_data
```
